#  seprate the name based on gender or first name or last name 

```python
#  ser name  name 
df['Ser_Name']=df["Name"].str.split(',').str.get(0)
```

```python
#  to add the gender miss,mr 
df['Gender']=df['Name'].str.split(',').str.get(1).str.strip().str.split(' ').str.get(0)
```

```python
#  add the 1st name 
df['F_Name']=df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1).str.get(1)
```
