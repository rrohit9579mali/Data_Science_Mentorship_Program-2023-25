```python
def team1_vs_team2(team1,team2):
    team_df=matches[(matches['Team1']==team1) & (matches['Team2']==team2) | (matches['Team2']==team1) & (matches['Team1']==team2)]
    match1_count=team_df['WinningTeam'].value_counts()[team1]
    match2_count=team_df['WinningTeam'].value_counts()[team2]
    totalmatch=team_df.shape[0]
    draw=totalmatch-(match1_count+match2_count)
    dic={
        'total_matches':int(totalmatch),
        team1:match1_count,
        team2:match2_count,
        'draw':draw
    }
    return dic
      
```
