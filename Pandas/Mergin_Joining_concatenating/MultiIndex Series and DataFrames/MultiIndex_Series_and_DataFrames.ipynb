```python
# can we have multiple index? Let's try
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
a = pd.Series([1,2,3,4,5,6,7,8],index=index_val)
a
```

```python
# The problem?
a['cse']
```

```python
# The solution -> multiindex series(also known as Hierarchical Indexing)
# multiple index levels within a single index
```

```python
# how to create multiindex object
# 1. pd.MultiIndex.from_tuples()
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex.levels[1]
# 2. pd.MultiIndex.from_product()
x=pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
```

```python
# creating a series with multiindex object
s = pd.Series([1,2,3,4,5,6,7,8],index=x)
s 
```

```python
s.unstack()
```

```python
death = death.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name='Date',
    value_name='Death'
)
# Optional: Only show the top rows for preview
death = death.head()

# Step 2: Melt the confirm DataFrame
confirm = confirm.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name='Date',
    value_name='confirm'
)

# Step 3: Merge both on the common columns
merged_df = confirm.merge(
    death,
    on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date']
)
```
