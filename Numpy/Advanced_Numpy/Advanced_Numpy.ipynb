```python
# broadcasting in numpy 
#  run for brodcasting 
# 1 dimension not math then convert the low dimension to high dimention  1-> 3D 2-> 1D 1D->3D
#  then 2nd arr convert into 3x3 is possible then brodcast either no posible

#  possible brodcast 
arr1=np.random.random(12).reshape(4,3)
arr2=np.random.random(3)
print(np.round(arr1+arr2))
```

```python
#  sigmoid function this is very famous and it is widly use in logistic machin learning alghorithum 
# formula of the sigmoid fn is 
#   1/(1+e^-x)

def sigmoid(array):
    return    1/(1+np.exp(-(array)))

a=np.random.random((4,3))
sigmoid(np.round(a))
```

```python
#  mean squared error 
#  yani hamara jo actual data and predicted data kitna k=jagalat hai vo predicat karata hai 
def MSE(actual,predicted):
    return np.mean(actual-predicted)**2
actual=np.random.randint(1,50,25)
predicted=np.random.randint(1,100,25)
ans=MSE(actual,predicted)
print(ans)
```

```python

# X      (0  2  100 )
x = np.linspace(0, 2*np.pi, 100)

# Y   sin(x)   
y = np.sin(x)

#   
plt.figure(figsize=(8, 5))  # Graph  Size  
plt.plot(x, y, label="Sine Wave", color="blue", linewidth=2)

#   Styling 
plt.title("Sine Wave Graph")  # Graph  
plt.xlabel("X values (radians)")  # X-axis 
plt.ylabel("sin(x) values")  # Y-axis 
plt.axhline(0, color='black', linewidth=1)  # X-axis  
plt.axvline(0, color='black', linewidth=1)  # Y-axis  
plt.grid(True, linestyle='--', alpha=0.6)  #   
plt.legend()  #  

#   
plt.show()
```
