### `Q-2:` Using `groupby` make groups using the `"Pclass"` column and fill every group's `"Embarked"` column's missing values with the mode value of that group. After that, print every group's `"Embarked"` column's value counts in ascending order.

````python
# code here
x=titanic_df['Embarked'].fillna(0)
q2=titanic_df.groupby('Pclass')['Embarked'].value_counts()
q2.reset_index()
````

````python
# Step 1: Define the mode-filler function
def fill_mode(x):
    return x.fillna(x.mode()[0])

# Step 2: Fill missing values with mode groupwise
titanic_df['Embarked'] = titanic_df.groupby('Pclass')['Embarked'].transform(fill_mode)

# Step 3: Print value counts in ascending order
for pclass, group in titanic_df.groupby('Pclass'):
    print(f"Pclass {pclass} Embarked counts:")
    print(group['Embarked'].value_counts().sort_values())
    print()   
````

### `Q-3:` Make groups based on `"Embarked"` column. And for each of this embarked group, make another group based on `"Pclass"` and find out the average fare (round off up to 2 decimal places) for each "Pclass" for each group of "Embarked".

**Sample Output:**

```bash
{'C': {1: 105, 2: 25, 3: 11},
 'Q': {1: 90, 2: 12, 3: 11},
 'S': {1: 70, 2: 20, 3: 15}}
```

````python
# code here
df=titanic_df.groupby(['Embarked','Pclass'])['Fare'].mean().round(2)
df.reset_index()
````
